```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv("spaceship-titanic/train.csv")
data.head()
```

```python
data.columns
```

```python
data.isnull().sum()
```

```python
data = data.dropna()
```

```python
data.isnull().sum()
```

```python
num_rows_before = len(data)
data = data.drop_duplicates(subset=['PassengerId', 'HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'Age',
       'VIP', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck',
       'Name', 'Transported'])
num_rows_after = len(data)
num_rows_deleted = num_rows_before - num_rows_after

print("Number of rows deleted:", num_rows_deleted)
```

```python
data['Transported_binary'] = data['Transported'].astype(int)
data['CryoSleep_binary'] = data['CryoSleep'].astype(int)
data['VIP_binary'] = data['VIP'].astype(int)
```

```python
data.head()
```

```python
non_numeric_columns = data.select_dtypes(exclude=['number']).columns
non_numeric_columns
```

```python
df= data.drop(['PassengerId', 'Name', 'Transported'], axis=1)
df = pd.get_dummies(df, columns=['HomePlanet', 'Cabin', 'Destination'])

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```
